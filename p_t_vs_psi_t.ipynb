```python


import numpy as np
import torch
import matplotlib.pyplot as plt
import torch.nn as nn
import torch.nn.functional as F
import random
import torch.optim.lr_scheduler as lr_scheduler

import numpy as np
import matplotlib.pyplot as plt
import torch

import math as m
import matplotlib.pyplot as plt
import matplotlib
from matplotlib.ticker import (MultipleLocator, AutoMinorLocator)
from matplotlib.ticker import FormatStrFormatter

plt.rcParams['text.usetex']=True
# plt.rcParams['text.latex.preamble']=r'\makeatletter \newcommand*{\rom}[1]{\expandafter\@slowromancap\romannumeral #1@} \makeatother'
# plt.rcParams['text.latex.preamble'] = [r'\usepackage{sfmath} \boldmath']
plt.rcParams['text.usetex'] = True
font = {'family' : 'normal',
        'weight' : 'bold',
        'size'   : 32}
matplotlib.rc('font', **font)



# Define the loss function
def loss_function(w1, w2):
    return (w1**2 - 4)**2 + (w2**2 - 1)**2

# Define the function for bm_psi(t, delta * w0)
def bm_psi(t, delta):
#     delta, w0_1, w0_2 = delta_w0
    term1 = (2 * delta) / np.sqrt(delta**2 + (8 - delta**2) * np.exp(-32 * t))
    term2 = delta / np.sqrt(delta**2 + (2 - delta**2) * np.exp(-8 * t))
    return term1, term2

# Define the function for rv_p(t)
def rv_p(t):
    return 2 / np.sqrt(1 + 4 * np.exp(-32 * t))

# Grid of (w1, w2) values
w1 = np.linspace(-0.5, 2.5, 100)
w2 = np.linspace(-0.5, 1.5, 100)
W1, W2 = np.meshgrid(w1, w2)
Z = loss_function(W1, W2)

# Plotting
fig, ax = plt.subplots(figsize=(15,10))
# Plot contour
contour = ax.contourf(W1, W2, Z, levels=50, cmap='Greys_r')
fig.colorbar(contour, ax=ax)

t_values = np.linspace(-1, 1, 100)
# Plot rv_p(t)
p_x = [rv_p(t) for t in t_values]
p_y = np.zeros_like(p_x)
ax.plot(p_x, p_y, color='lightgreen', linestyle='-', label=r'$p(t)$',linewidth = 2)
ax.set_xlabel('$w_1$')
ax.set_ylabel('$w_2$')


dlt_all = [0.1,0.05,0.001]
ls = ['-','--',':']
for i in range(3):
    # Overlay the curves
    delta = dlt_all[i]
    t_values = np.linspace(0, 3, 100)


    # Plot bm_psi(t, delta * w0)
    w0 = (1 / np.sqrt(2), 1 / np.sqrt(2))
    psi_x = []
    psi_y = []
    for t in t_values:
        psi = bm_psi(t, delta)
        psi_x.append(psi[0])
        psi_y.append(psi[1])
    ax.plot(psi_x, psi_y, color='red', label=r'$\delta = $ '+str(delta),linestyle=ls[i],linewidth = 2)



    # Add labels and legend

#     ax.set_title('Contour Plot of Loss Function with Curves')
    ax.legend()
plt.scatter(2,0,c='lightgreen')
plt.scatter(2,1,c='red')
plt.show()
```
